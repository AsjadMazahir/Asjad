## Discretization using sklearn

```python
import pandas as pd
import numpy as np
import seaborn as sns

# loading dataset
df = sns.load_dataset('titanic')
df.head()
```

```python
# imputing missing values
df['age'] = df['age'].fillna(df['age'].median())
df['fare'] = df['fare'].fillna(df['fare'].median())
```

```python
from sklearn.preprocessing import KBinsDiscretizer
age_dicret = KBinsDiscretizer(n_bins=5, encode='ordinal', strategy='uniform')
df['age_bin'] = age_dicret.fit_transform(df[['age']])
df.head()
```

```python
sns.histplot(df, x='age', hue='age_bin')
```

## Binning method with pandas

```python
# pandas method for manual binning
df['age_bin2'] = pd.cut(df['age'], bins=5, labels = ['child', 'teen', 'young', 'adult', 'old'])
df.head()
```

```python
bins = [0, 10, 18, 30, 50, 80]
labels = ['child', 'teen', 'young', 'adult', 'old']
df['age_bins3'] = pd.cut(df['age'], bins=bins, labels=labels)
sns.histplot(df, x='age', hue='age_bins3')
```
