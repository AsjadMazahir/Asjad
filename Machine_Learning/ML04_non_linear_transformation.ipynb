## Non linear Transformation

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# generate normal data
np.random.seed(0)
df = pd.DataFrame(np.random.exponential(size=1000, scale=2), columns=['values'])
df.head()
```

```python
# visualize the data
sns.histplot(df, kde=True)
```

```python
from sklearn.preprocessing import PowerTransformer
from sklearn.preprocessing import QuantileTransformer

pt_box_cox = PowerTransformer(method='box-cox', standardize=True)
pt_yeo_johnson = PowerTransformer(method='yeo-johnson', standardize=True)
pt_quantile = QuantileTransformer(output_distribution='normal')

df['Box-Cox'] = pt_box_cox.fit_transform(df[['values']] + 1)
df['Yeo-Johnson'] = pt_yeo_johnson.fit_transform(df[['values']])
df['Quantile_Transform'] = pt_quantile.fit_transform(df[['values']])
df.head()
```

```python
for col in df.columns:
    sns.histplot(df[col], kde=True)
    plt.show()
```

## L1 and L2 Normalization

```python
from sklearn.preprocessing import Normalizer
data = [[1,1,1], [1,1,0], [1,0,0]]
# normalize the data
norm = Normalizer(norm='l1')
print(norm.fit_transform(data))
```

## Log transformation

```python
import pandas as pd
import numpy as np

# example dataset with skewed values
df = {'Values': [1, 4, 6, 10, 20, 30, 50, 100, 100, 200, 500, 700, 1000, 2000, 3000, 5000, 6000, 10000, 200000, 500000]}
df = pd.DataFrame(df)
df.head()
```

```python
df['log_vales'] = np.log(df['Values'])
df
```

```python
for col in df.columns:
    sns.histplot(df[col], kde=True)
    plt.show()
```
