## Correlation

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = sns.load_dataset('titanic')
```

```python
sns.heatmap(df[['age', 'fare']].corr(), annot=True)
```

```python
# correlation on tips dataset
df2 = sns.load_dataset('tips')
df2.head()
```

```python
sns.lmplot(data=df2, x='total_bill', y='tip')
```

```python
# computing correlation
df2['tip'].corr(df2['total_bill'])
```

```python
sns.pairplot(df2)
```
