```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Important things about plotting according to grammar of graphics

2. Types of plots
3. x axis
4. y axis
5. grouping (hue/color) based on categorical variable
6. size of the plot
7. color palette of the plot
8. style of the plot
9. interpretation

```python
# loading data
df = sns.load_dataset('titanic')
```

### Scatter Plot

```python
plt.figure(figsize=(8,4))
sns.set_theme(style='darkgrid', font_scale=1.2)
palette = {'male': 'blue', 'female': 'red'}
sns.scatterplot(data=df, x='age', y='fare', hue='sex', palette=palette)
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title("Fare vs Age scatter plot")
```

### Box plot

```python
plt.figure(figsize=(8,4))
sns.set_theme(style='darkgrid', font_scale=1.2)
palette = {'First': 'blue', 'Second': 'red', 'Third': 'green'}
sns.boxplot(data=df, x='sex', y='age', hue='class', palette=palette)
plt.xlabel('Sex')
plt.ylabel('Age')
plt.title("Sex vs Age box plot")
plt.legend(loc='upper left', title='Class Name')
```

### Bar Plot

```python
plt.figure(figsize=(8,4))
sns.set_theme(style='darkgrid', font_scale=1.2)
palette = {'First': 'blue', 'Second': 'red', 'Third':'yellow'}
sns.barplot(data=df, x='sex', y='fare', hue='class', palette=palette, capsize=0.08 )
plt.xlabel('Sex')
plt.ylabel('Fare')
plt.title("Fare vs Sex bar plot")
```
