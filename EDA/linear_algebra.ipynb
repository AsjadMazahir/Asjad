## Linear Algebra
### Solving system of linear equation
2x + 3y =8 \
5x + 4y =20

```python
import numpy as np
```

```python
# coefficient of matrix A
A = np.array([[2, 3], [5, 4]])
print(A)
# constant matrix B
B = np.array([8, 20])
print(B)
```

```python
# Solve for x
x = np.linalg.solve(A, B)
x
```

## Solving complex equations

```python
A_complex = np.array([[1+2j, 2-1j], [3-1j, 4]])
B_complex = np.array([4+1j, 3+4j])
x_complex = np.linalg.solve(A_complex, B_complex)

x_complex
```

## Vectors in python

```python
import matplotlib.pyplot as plt
v = np.array([3,5])
v
```

```python
plt.figure()
plt.quiver(0, 0, v[0], v[1], angles = 'xy', scale_units='xy', scale=1, color='r')
plt.xlim(0,7)
plt.ylim(0,7)
plt.grid()
plt.title('2D vector')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```

## Linear Transformation

```python
# Define the rotation matrix for 45 rotation
theta = np.radians(45)
R = np.array([[np.cos(theta), -np.sin(theta)], 
              [np.sin(theta), np.cos(theta)]])
print(R)
print("--------------")
print(v)
v_trans = np.dot(R, v)
v_trans
```

```python
plt.figure()
plt.quiver(0, 0, v[0], v[1], angles = 'xy', scale_units='xy', scale=1, color='r', label='Original Vector')
plt.quiver(0, 0, v_trans[0], v_trans[1], angles = 'xy', scale_units='xy', scale=1, color='b', label='Transformed Vector')
plt.xlim(-2,7)
plt.ylim(-2,7)
plt.grid()
plt.title('2D vector transformation')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.legend()
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```

## SVD in python

```python
# generate a matrix of 4 x 3 
A = np.random.rand(4,3)
A
```

```python
# lets apply SVD
U, S, V = np.linalg.svd(A)
print("U: ", U)
print("----------------------")
print("S: ", S)
print("----------------------")
print("VT: ", V)
```
