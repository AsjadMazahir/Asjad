## Data Visualization Part 1

```python
# importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# loading titanic dataset
df = sns.load_dataset('titanic')
df.head()
```

```python
df.plot(kind='scatter', x='age', y='fare')
plt.title('Fare vs age plot')
```

```python
df.groupby('survived')['fare'].mean().plot(kind='bar')
```

```python
# bar plot on df using seaborn
sns.barplot(df, x='survived', y='fare', hue='class')
```

```python
# scatter plot using sns
sns.scatterplot(df, x='age', y='fare', hue='class')
```

```python

sns.scatterplot(df, x='age', y='fare', hue='survived')
```
