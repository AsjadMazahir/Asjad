```python
# importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
df = sns.load_dataset('tips')
df.head()
```

### Drawing a hist plot
---

```python
sns.histplot(data=df, x='total_bill', kde=True)
```

```python
sns.histplot(data=df, x='tip', kde=True)
```

```python
df1 = sns.load_dataset('titanic')
df1.head()
```

```python
sns.histplot(data=df1, x='age', kde='True', hue='class')
```

```python

sns.boxplot(data=df1, x='sex', y='age', hue='class')
```

## Kurtosis and Skewness
---

```python
# finding kurtosis of age
stats.kurtosis(df1['fare'], fisher=False)
```

```python
stats.skew(df1['fare'])
```

```python
np.random.seed(0)
data = np.random.normal(0,1,1000)
data_skewness = stats.skew(data)
data_kurtosis = stats.kurtosis(data, fisher=False)
print(f"Skewness of data:  {data_skewness}")
print(f"Kurtosis of data: {data_kurtosis}")
```

```python
sns.histplot(data, kde=True)
```
